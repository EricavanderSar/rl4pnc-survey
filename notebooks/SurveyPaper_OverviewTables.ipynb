```python
measures = ['agent_type', 
            'opponent', 
            # 'action space', 
            'completed episodes',
            'steps survived', 
            'steps overloaded', 
            'agent execution time [ms]',
            'maximum topology depth',
            'unique actions',
            'unique lines in danger',
            'unqique subs changed',
            'substations changed'
           ]
filter_dn_base = ((df_heur['agent_type'] == "DoNothing") & ~(df_heur['line_reco']) & ~(df_heur['line_disc']))
filter_greedy_base = ((df_heur['agent_type'] == "RhoGreedy") & (df_heur['action space'] == "medha") & (df_heur['line_reco']) & ~(df_heur['line_disc']) & (df_heur['reset_topo']==2.0) )
base_heur = df_heur.loc[(filter_dn_base | filter_greedy_base), measures]
base_heur
```

```python
measures = ['agent_type', 
            'opponent', 
            'train_opponent',
            # 'action space', 
            'completed episodes',
            'steps survived', 
            'steps overloaded', 
            'agent execution time [ms]',
            'maximum topology depth',
            'unique actions',
            'unique lines in danger',
            'unqique subs changed',
            'substations changed'
           ]
filter_rl_base = ((df_rl['action space'] == "medha") & (df_rl['line_reco']) & ~(df_rl['line_disc']) & (df_rl['reset_topo']==0.0) )

base_rl = df_rl.loc[filter_rl_base, measures]
base_rl
```

```python
base_rl.groupby(['opponent', 'train_opponent']).mean()
```

```python
all_subs_changed = []
for subs in base_rl.loc[((base_rl['opponent']==False) & (base_rl['train_opponent']==False))]['substations changed']:
    subs = subs[1:-1].split()
    for sub in subs:
        if int(sub) not in all_subs_changed:
            all_subs_changed.append(int(sub))

all_subs_changed.sort()
all_subs_changed
```
