# Collect action spaces results

```python
measures = ['agent_type', 
            'opponent', 
            'action space', 
            'completed episodes',
            'steps survived', 
            'steps overloaded', 
            'agent execution time [ms]',
            'maximum topology depth',
            'unique actions',
            'unique lines in danger',
            'unqique subs changed',
            'substations changed'
           ]
filter_greedy_actspace = ((df_heur['agent_type'] == "RhoGreedy") & (df_heur['action space'].isin(["medha","tennet","assym","d3qn2022"])) & (df_heur['line_reco']) & ~(df_heur['line_disc']) & (df_heur['reset_topo'].isin([0.0,2.0])) )
base_heur = df_heur.loc[(filter_greedy_actspace), measures]
base_heur
```

```python
measures = ['agent_type', 
            'opponent', 
            'train_opponent',
            'action space', 
            'completed episodes',
            'steps survived', 
            'steps overloaded', 
            'agent execution time [ms]',
            'maximum topology depth',
            'unique actions',
            'unique lines in danger',
            'unqique subs changed',
            'substations changed'
           ]
df_rl.loc[:, measures].groupby(['opponent', 'train_opponent', 'action space']).mean()
```

```python
all_subs_changed = []
act_spaces = df_rl['action space'].unique()

for act_space, opp, train_opp in itertools.product(act_spaces, [False,True], [False,True]):
    all_subs_changed = []
    filter = ((df_rl['action space'] == act_space) & (df_rl['opponent'] == opp)& (df_rl['train_opponent'] == train_opp))
    for subs in df_rl.loc[filter]['substations changed']:
        subs = subs[1:-1].split()
        for sub in subs:
            if int(sub) not in all_subs_changed:
                all_subs_changed.append(int(sub))
    all_subs_changed.sort()
    print(f"Substations changed {all_subs_changed} for action space: {act_space}, Opponent: {opp}, Train Opp:{train_opp}")
```

```python
path = "/Users/ericavandersar/surfdrive/Documents/Research/Results_SurveyPaper/"

file_name_summ_data = "ObsExp.csv"

rl_path = os.path.join(path, "PPO")
df_rl = pd.read_csv(os.path.join(rl_path, file_name_summ_data))
df_rl = df_rl.sort_values(['opponent','train_opponent'], ascending=[True, True])
df_rl
```

```python
obs_path = os.path.join(path, "PPO")
df_obs = pd.read_csv(os.path.join(obs_path, "Obs_spaces_agent_ids.csv"))
df_obs
```

```python
# Create a mapping from agent_id to observation space in df_obs
obs_mapping = df_obs.set_index('agent_id')['observation space']

# Update the "observation space" column in df_rl based on the mapping
df_rl['observation space'] = df_rl['agent_id'].map(obs_mapping).combine_first(df_rl['observation space'])

# Display the updated df_rl
df_rl
```

```python
df_rl = df_rl.sort_values(['opponent','train_opponent', 'observation space'], ascending=[True, True, True])
df_rl
```

```python
df_rl = df_rl.dropna(subset=['observation space'])
```
